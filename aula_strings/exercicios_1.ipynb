```python
## Comprimento da String

a = "Olá Lucas Luan, bem vindo ao curso de Python!"
print(len(a))
```

```python
## Inversão de string

a = "Hello My Friend, I'm not good today, I'm Tireddddd!"
print(a[::-1])
```

```python
## Contagem de vogiais

a = "O Rato roeu a roupa do Rei de Roma"
vogais = "aeiouAEIOU"
contador = 0
for letra in a:
    if letra in vogais:
        contador += 1      

print(f"A string contém {contador} vogais.")
```

```python
## Remoção de vogais e consoantes

a = "Hoje esbarrei num paralelepípedo e quase caí!"
vogais = "aeiouAEIOU"
consoantes = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
sem_vogais = ""
sem_consoantes = ""
for letra in a:
    if letra in vogais:
        sem_consoantes += letra
    elif letra in consoantes:
        sem_vogais += letra
    else:
        sem_vogais += letra
        sem_consoantes += letra

print(f"String sem vogais: {sem_vogais}")
print(f"String sem consoantes: {sem_consoantes}")
```

```python
## Remoção de espaços extras

a = "   Olá Lucas Luan, bem vindo ao curso de Python!   "
a = a.replace("   ", "")
print(a)
```

```python
## Verificação de palíndromo

a = "Arara"
a = a.lower()
if a == a[::-1]:
    print("A string é um palíndromo.")
else:
    print("A string não é um palíndromo.")
```

```python
## Tokenização simples

a = "Olá Lucas Luan, como você veio ao curso?"
tokens = a.split()
print(tokens)
```

```python
## Substituição de caracteres

a = "Olá Lucas Luan, bem vindo ao curso de Python!"
a = a.replace("o", "0").replace("a", "4").replace("cur", "tor")
print(a)
```

```python
## Extração de iniciais

a = "João Mineiro e Marciano"
b = a.split()
iniciais = ""

for nome in b:
    iniciais+= nome[0] + "."
print(iniciais)
```

```python
a = "Oi, meu nome é Lucas Luan Nunes, formado em Ciência de dados e IA pela UniDomBosco, faço Pós graduação EM IA Generativa na Universidade Federal do Paraná e também faço Pós Graduação de Engenharia de Dados e IA com Foco na Industria 4.0 na PUCPR"

n = 20
chunks = []
i = 0

while i < len(a):
    fim = i + n

    if fim >= len(a):
        chunks.append(a[i:].strip())
        break

    # Se o corte cair no meio de uma palavra, volta até o último espaço
    if a[fim] != " ":
        ultimo_espaco = a.rfind(" ", i, fim)
        if ultimo_espaco != -1 and ultimo_espaco > i:
            fim = ultimo_espaco

    chunk = a[i:fim].strip()
    chunks.append(chunk)

    i = fim

    # pula espaços entre os chunks
    while i < len(a) and a[i] == " ":
        i += 1

print(chunks)
```

```python
# Contagem de frequência de palavras

a = "Quem casa quer casa, mas quem não tem casa não casa, porque quem não tem casa não tem casa que casa quem quer casar."
palavras = a.split()
frequencia = {}
for palavra in palavras:
    palavra = palavra.strip(",.").lower()
    if palavra in frequencia:
        frequencia[palavra] += 1
    else:
        frequencia[palavra] = 1
print(frequencia)   
```
