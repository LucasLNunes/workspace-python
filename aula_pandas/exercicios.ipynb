```python
# Manipulação de strings e categorização: Dataset `Conversations (Português)`: Crie uma
# coluna `prompt_length` com o número de caracteres do texto do usuário, outra com o texto em
# caixa-alta (`str.upper()`). (https://huggingface.co/datasets/nicholasKluge/instruct-aira-dataset-
# v2)



dataset = load_dataset(
    "nicholasKluge/instruct-aira-dataset-v2",
    split="portuguese"
)

df = dataset.to_pandas()
df.head()
```

```python
# Manipulação de strings e categorização: Dataset `Conversations (Português)`: Crie uma
# coluna `prompt_length` com o número de caracteres do texto do usuário, outra com o texto em
# caixa-alta (`str.upper()`). (https://huggingface.co/datasets/nicholasKluge/instruct-aira-dataset-
# v2)

df['user_text'] = df['conversations'].apply(lambda x: x[0]['content'])
df['prompt_length'] = df['user_text'].str.len()
df['user_text_upper'] = df['user_text'].str.upper()
df[['user_text', 'prompt_length', 'user_text_upper']].head()
```

```python
# Aplicando funções com `apply()`: Dataset `prompts.csv`: Escreva uma função que classifica
# prompts como “curto” (<20 chars), “médio” ou “longo”, e aplique usando `df['prompt'].apply()`

def classify_prompt_length(prompt):
    length = len(prompt)
    if length < 20:
        return 'curto'
    elif length < 50:
        return 'médio'
    else:
        return 'longo'

df1['length_category'] = df1['prompt'].apply(classify_prompt_length)
```
